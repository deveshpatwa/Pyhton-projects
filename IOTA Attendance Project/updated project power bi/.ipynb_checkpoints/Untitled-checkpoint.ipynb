```python
import pandas as pd
```

```python
leave_data = pd.read_excel("leave_data.xlsx",sheet_name='leave')

# selecting only necessory column
leave_data =  leave_data[['Name','Leave Start Date','Leave End Date','Half Day/ Full Day?','Number of Days']]
```

```python
leave_data
```

```python
a = leave_data.to_dict()
```

```python
a.keys()
```

```python
length_of_dict = len(a['Name'])
```

```python
data = pd.DataFrame()
```

```python
temp_data = pd.DataFrame()
for i in range(length_of_dict):
    try :
        all_dates = pd.DataFrame(pd.date_range(start=a['Leave Start Date'][i].dt.date,end=a['Leave End Date'][i].dt.date))
        all_dates['Half Day/ Full Day'] = a['Half Day/ Full Day?'][i]
        all_dates['Name'] = a['Name'][i]
        pd.concat((temp_data,all_dates))
    except:
        pass
temp_data
```
