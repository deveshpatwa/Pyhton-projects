```python
def popsample(sz,sample):
    a=[]
    for i in range(sz):
        a.append(s.sample(sample).mean())
    a = pd.Series(a)
    plt.subplot(1,2,1)
    pd.Series(a).plot(kind='hist')
    plt.subplot(1,2,2)
    s.plot(kind='hist')
    print('population mean',s.mean())
    print('sample mean',a.mean())
```

```python
popsample(100,30)
```
