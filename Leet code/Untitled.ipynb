```python
text = "hello world"
broken_letters ="ad"

num = 0
for i in text.split():
    temp = 0
    for j in i:
        if j in broken_letters:
            temp += 1
            
    if temp ==0:
        num +=1
print(num)
```

```python
num = [2,7,11,15] 
target = 26

def findundex(num,target):
    for i in num:
        temp = target - i
        if temp in num:
            return [num.index(i),num.index(temp)]

findundex(num,target)
```

```python
num = [2,7,11,15] 
target = 9

len_of_list = len(num)

for i in range(len_of_list):
    for j in range(i+1,len_of_list):
        if num[i]+num[j] == target:
            print([i,j])
```
