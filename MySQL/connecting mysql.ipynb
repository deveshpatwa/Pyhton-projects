```python
# importing mysql
import mysql.connector
import pandas as pd
from sqlalchemy import create_engine
```

you can use to **connect with any database** by using - "mysql+mysqlconnector://**username:password@localhost/dbname**"

```python
user_name = "root"
password = "1234"
server = "localhost"
database = "db1"
engine = create_engine(f"mysql+mysqlconnector://{user_name}:{password}@{server}/{database}")
```

```python
def run(query):
    df = pd.read_sql(query, con=engine)
    return df
```

```python
run("select * from store where sale<=500")
```

```python
run("SHOW TAbles")
```

```python
run("select * from sale_year")
```

```python
run("select * from store where category = 'furniture'")
```

```python
run("drop table emp")
run("create table emp (name varchar(40),dep varchar(30),salary int)")
```
