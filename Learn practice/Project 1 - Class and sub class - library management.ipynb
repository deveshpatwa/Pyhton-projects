## Base class

```python
# 1. Base class - parent class
class library_items:
    def __init__(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
        # print("--employee has been created--")

    def display_info(self):
        # print("-- displaying info of an employee")
        print("Title:",self.title)
        print("Author:",self.author)
        print("Year:",self.year)
        # print("--till here its parents method")
```

## Create two subclasses Book and Magazine

```python
class book(library_items):
    """
    Hello this class will be managinf books in library
    """
    def __init__(self,title,author,year,genere):
        super().__init__(title,author,year)
        # print("--just ran the init from parent")
        self.genere = genere
        # print("-- now iam running child specific method")
        

# now creating method overriding the dispplay menthon of parent
    def desplay_info(self):
        super().display_info()
        print("genere:",self.genere)
        

# Magazine sub class
class magazine(library_items):
     def __init__(self,title,author,year,issue_number):
         super.__init__(self,title,author,year)
         self.issue_number = issue_number
         
```

```python
class ebook(book):
    def __init__(self,title,author,year,genere,file_size):
        super().__init__(title,author,year,genere)
        self.file_size = file_size

    def display_info(self):
        super().display_info()
        print("File size:",self.file_size)
```
