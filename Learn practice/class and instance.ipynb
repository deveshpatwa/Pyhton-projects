## Bank Account Class
Create a BankAccount class with attributes account_number, balance, and methods deposit(amount) and withdraw(amount). Write a method to display the account balance.

```python
class bank_account:
    start_account = 10001
    def __init__(self,balance=0):
        self.balance = balance
        self.account_number = bank_account.start_account
        bank_account.start_account += 1

    def deposit(self,amount):
        self.balance += amount

    def withdraw(self,amount):
        self.balance -= amount

    def check_balance(self):
        return self.balance
        
```

## Library System
Design a Book class with attributes title, author, and copies. Add methods borrow_book() and return_book(). Ensure that a book cannot be borrowed if no copies are left.

```python
class book:
    def __init__(self,title,author="Unknown",copies=1):
        self.title = title.title()
        self.author = author.title()
        self.copies = copies

# borrowing a book method
    def borrow_book(self):
        if self.copies == 0:
            return "No books left !!"
        else:
            self.copies -= 1
            return "Book has been allotted"

# now method for returning a book
    def return_book(self):
        self.copies += 1

# seeling all the info of book method
    def info(self):
        print("Title :",self.title)
        print("Author :",self.author)
        print("Number of copies lef :",self.copies)
```

```python
morning = book("morning","devesh",2)
```

```python
morning.info()
```

```python
morning.borrow_book()
```

## Employee Management
Build an Employee class with attributes name, id, and salary. Write a method to increase salary by a given percentage.


```python
class employee:
    base_increment = 0.098
    emp_id = 101

    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
        self.id = employee.emp_id
        employee.emp_id += 1

    def salary_increment(self,increment=base_increment):
        self.salary = self.salary * (1+increment)
```

## Student Grade System
Make a Student class with attributes name and a dictionary of subjects with marks. Add methods to calculate average marks and display grade (A, B, C, etc.) based on average.

```python
class student:
    def __init__(self,name,subjects):
        self.name = name
        self.subjects = subjects
    def average_marks(self):
        total = 0
        marks = self.subjects.values()
        for i in marks:
            total += i
        return total/len(marks)
```

```python
s1 = student("ravi",{"maths":56,"physics":99})
```

```python
s1.average_marks()
```

## Inheritance in python
creating sub-class of developer which is an employee of an oreganisation using the main class employee and parent and sub class of developer as child 

```python
class employee:
    base_increment = 0.098
    emp_id = 101

    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
        self.id = employee.emp_id
        employee.emp_id += 1

    def salary_increment(self,increment=base_increment):
        self.salary = self.salary * (1+increment)

class developer(employee):
    def __init__(self,name,salary,prog_language):
        super().__init__(name,salary)
        self.prog_language = prog_language
        
```
