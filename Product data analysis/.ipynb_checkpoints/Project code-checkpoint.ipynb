# Importign Library

```python
import pandas as pd
```

# Loading Data File

```python
df = pd.read_csv("Product Sales Data.csv")
```

```python
df.head()
```

# Data Cleaning

```python
df.info()
```

```python
df['Sales'].astype(int)
```

```python
def clean_amount(num):
    return int(num.replace(",",""))
```

```python
clean_amount("4,28,700")
```

```python
df['Sales'] = df['Sales'].apply(clean_amount)
```

```python
df['Qty'] = df['Qty'].apply(clean_amount)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Sub-Channels'] = df['Sub-Channels'].apply(lambda x: x.title())
```

```python
df['Product Name'] = df['Product Name'].apply(lambda x: x.title())
```

```python
df.head()
```

```python
df['Month'].unique()
```

```python
df['date'] = pd.to_datetime(df['Month'])
del df['Month']
```

```python
df.head()
```

# Data Analysis questions

## Q1. Find Category wise total sales 

```python
df.groupby("Category")['Sales'].sum() 
```

## Q2. Find top 5 most sold products
ignore the packaging quantity

```python
df.groupby("Product Name")['Sales'].sum()
```

```python
def clean_name(text):
    for i in text:
        if i.isdigit():
            return text.split(i)[0].strip()
```

```python
clean_name("Virgin Coconut Oil 250 Ml ")
```

```python
df['product name only'] =  df['Product Name'].apply(clean_name)
```

df.groupby("product name only")['Sales'].sum()

## Q3. Which month we had the heighest total sales

```python
df.head()
```

```python
df['month name'] = df['date'].dt.month_name()
```

```python
df.groupby("month name")['Sales'].sum().sort_values(ascending=False).head(1)
```

## Q4. Top 3 products in each category

```python
aggdf = df.groupby(["Category",'product name only'])['Sales'].sum()
aggdf
```

```python
aggdf = pd.DataFrame(aggdf).reset_index()
aggdf
```

```python
aggdf['rank'] = aggdf.groupby('Category')['Sales'].rank(ascending=False)
aggdf
```

```python
aggdf = aggdf[aggdf['rank'] <= 3]
aggdf
```

```python
aggdf = aggdf.sort_values(by=['Category',"Sales"], ascending=[True,False]).drop(columns='rank')
```

```python
aggdf.to_csv("final report.csv")
```

## Q5. Channel and subchannel wise total number of orders 

```python
df.groupby(['Channel',"Sub-Channels"])['date'].count()\
                    .reset_index().rename(columns={'date':"Number of Orders"})\
                    .sort_values(by=['Channel',"Number of Orders"],ascending=False)
```
