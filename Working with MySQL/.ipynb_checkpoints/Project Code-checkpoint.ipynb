# way 1 to do this using MySQL connector

```python
import mysql.connector as mysql
import pandas as pd
```

```python
build_connection = mysql.Connect(
    host="localhost",
    user = "root",
    database= 'db1',
    password='1234'
)
```

```python
cursor = build_connection.cursor()
```

```python
pd.read_sql("select * from store",build_connection)
```

# way 2 using SQLAlachemy
best if you want to build industri level projects

```python
import sqlalchemy as sql
```

```python
engine = sql.create_engine("mysql+mysqlconnector://root:1234@localhost/db1")
```

```python
pd.read_sql("select * from store;",engine)
```
