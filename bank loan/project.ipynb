```python
def uni_analysis_numerical(df,col):
    sns.set(style='darkgrid')
    plt.figure(figsize=(25,5))
    plt.subplot(1,3,1)
    sns.boxplot(data=df,x=col,orient='v').set(title='Box plot')
    plt.subplot(1,3,2)
    sns.histplot(df[col].dropna(),kde=True).set(title='Histogarm')
    plt.show()
```

```python
 uni_analysis_numerical(df_app,'AGE_IN_YEARS')
```

```python
def uni_analysis_categorical(df,col):
    sns.set(style='whitegrid')
    plt.figure(figsize=(12,5))
    df[col].value_counts().plot.barh(width=0.8, color='skyblue')
    plt.title(f'ferquenct of {col}')
    plt.show()
```
