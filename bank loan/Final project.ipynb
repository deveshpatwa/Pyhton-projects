```python
def ro(df,col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1                   # inter quirtile range quater 3 minus quater 1
    lb = q1 - 1.5*iqr                 # lower bound
    ub = q3 + 1.5*iqr                     # upper bound
    return df.loc[(df[col]>lb)&(df[col]<ub)]
```
