## What is Central Limit Theorem

#### The central limit theorem (CLT) states that, under appropriate conditions, the distribution of a normalized version of the sample mean converges to a standard normal distribution. This holds even if the original variables themselves are not normally distributed. 

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data in df
df = pd.read_csv("medical_examination.csv")
df = df[['Id','Weight']]
```

```python
# getting population mean
print(f"Size of Population = {len(df):,}")
print("Population Mean Weight = ",df['Weight'].mean())

sns.histplot(x=df['Weight'],bins=30)
plt.title("Weight distribution of weight of population")
plt.xlabel("Weight in KG.")
```

```python
# checking sample mean of 50 people
print(round(df.sample(50)['Weight'].mean(),2))
```

```python
# creating a list of sample mean by taking 100 sample
sample = []
for i in range(1000):
    sample.append(df.sample(50)['Weight'].mean())
sample = pd.Series(sample)
```

```python
print("Median weight of sample = ",sample.median())
sns.histplot(x=sample,bins=30)
plt.title("Frequency distribution on sample mean")
plt.xlabel("Weight in KG")
plt.ylabel("Frequency")
```
