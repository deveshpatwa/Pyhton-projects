```python
import numpy as np 
import pandas as pd
import seaborn as sns
```

```python
def compound(amount=10000,rate=0.08,year=1,period=1):
    return (amount*((1+(rate/period))**(year*period)))-amount
```

```python
def effect(num):
    series = pd.Series([compound(period=i) for i in range(1,num+1)])
    print(series)
    return sns.lineplot(series)
```

```python
effect(20)
```
