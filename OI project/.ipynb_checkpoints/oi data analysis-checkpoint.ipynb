```python
import pandas as pd
import requests as r
from bs4 import BeautifulSoup as bs
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://www.nseindia.com/api/equity-stockIndices?index=NIFTY%2050"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept": "application/json",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.nseindia.com/"
}
```

```python
session = r.Session()
session.get("https://www.nseindia.com", headers=headers)  # warm-up request
```

```python
response = session.get(url, headers=headers)
data = response.json()
```

```python
data['data'][0]['lastPrice']
```

```python
nifty_price = data["data"][0]["lastPrice"]
print("Latest NIFTY 50 Price:", nifty_price)
```

```python
def get_nifty_atm(price):
    if price%100 >= 50:
        return price//100 * 100 +100
    else:
        return price//100 * 100
```

```python
nifty_price = get_nifty_atm(nifty_price)
```

```python
def topn(df,nifty_price,num=5 ):
    num *= 100
    start = nifty_price - num
    end = nifty_price + num
    return df[df['strike'].between(start,end)]
```

## getting new way

```python
data = pd.read_html("datafile.html")
```

```python
df = data[1]
```

```python
columns = df.columns.to_list()
```

```python
new_columns = []
for i in columns:
    new_columns.append((i[0]+"_" +i[1]).lower())
```

```python
df.columns =  new_columns
```

```python
df.columns
```

```python
df = df[['calls_oi','calls_ltp','unnamed: 11_level_0_strike','puts_ltp', 'puts_oi']].rename(columns={"unnamed: 11_level_0_strike":"strike"})
```

```python
df['pcr'] = df['puts_oi'] / df['calls_oi']
```

```python
df.head()
```

```python
print("Put call ration - ",end="")
print((df['puts_oi'].sum() / df['calls_oi'].sum()).round(2),"%")
```

```python
df = df[df['strike']%100==0]
```

```python
print("Nifty current strike = ",nifty_price)
plot_data = topn(df,nifty_price)
plot_data
```

```python
for i in [1,2,3,4]:
    a = topn(df,nifty_price,i)
    pcr = (a['puts_oi'].sum() / a['calls_oi'].sum()).round(2)
    print(i*2+1, "Strike price PCR  -",pcr,"%" )
```

```python
plot_data.melt(value_vars=['calls_oi',"puts_oi"])
```

```python
plot_data[['calls_oi',"puts_oi"]].plot(kind="bar")
```
