```python
# This is called stratified shuffle split 
df['median_income'] = np.ceil(df['median_income'] /1.5 )
df['median_income'] = df['median_income'].where(df['median_income']<5,5)
```

```python
def find_numeric_col_names(df):
    l = []
    for i in df.columns:
        if df[i].dtype == 'int64' or df[i].dtype == float:
            l.append(i)
    return l
```

```python
l = []
for i in df.columns:
    if df[i].dtype == 'int64' or df[i].dtype == float:
        l.append(i)
a = df[l].corr()['median_house_value'].sort_values(ascending=False)
```

```python
df['bedroom_per_peson'] = df['total_bedrooms'] / df['households']
```

```python
df['rooms per person'] =  df['total_rooms'] / df['population']
```

```python
df[find_numeric_col_names(df)].corr()['rooms per person'].sort_values(ascending=False)
```

```python
df[find_numeric_col_names(df)].corr()['bedroom_per_peson'].sort_values(ascending=False)
```

```python
df.isnull().sum()
```

```python
for i in find_numeric_col_names(df):
    if df[i].isnull().sum() != 0:
        df[i].fillna(df[i].median(),inplace=True)
```
